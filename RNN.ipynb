```python
import os
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
os.chdir(r"D:\Imarticus\Time Series And Deep Learning\NLP")

amazon=pd.read_csv("amazonreviews.tsv",sep="\t")

amazon  
```

```python
amazon.label.replace({"pos":1,"neg":0},inplace=True)
```

```python
amazon.head()
```

```python
amazon.label.value_counts()
```

```python
amazon_x=amazon.iloc[:,1]
amazon_y=amazon.iloc[:,0]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(amazon_x,amazon_y,test_size=0.2)
```

```python
from keras.utils.np_utils import to_categorical
from tensorflow.keras.utils import to_categorical

y_train= to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
y_test
```

```python
## HyperParameters 
max_num_words=1200    ## How many words do we consider from the entire CORPUS(Coll of doc)
seq_len =100          ## how many token (Word to consider in each document/message)
embedding_size= 100   ## Vector length of each word
```

```python
from keras_preprocessing.text import Tokenizer  # Is used for tokinization
from keras_preprocessing.sequence import pad_sequences # Used to bring all samples to same length
```

### <font color="red">What is Tokenizer and pad_sequences

```python
tokenizer= Tokenizer(num_words=max_num_words)
tokenizer.fit_on_texts(amazon.review)     ## Would be used to assign the token
x_train=tokenizer.texts_to_sequences(x_train) ## Will convert the text to sequence of IDs
x_train=pad_sequences(x_train,maxlen=seq_len)
```

```python
x_train.shape
```

```python
x_train[1]
```

```python
x_test=tokenizer.texts_to_sequences(x_test) 
x_test=pad_sequences(x_test,maxlen=seq_len)
```

```python
from keras.models import Model,Sequential
from keras.layers import Dense,LSTM,Embedding
from tensorflow.keras.optimizers import Adam
```

```python
model=Sequential()  # Initialize the network
model.add(Embedding(input_dim=max_num_words, ## Embedding takes input in sequence and gives output in embedding
                   input_length= seq_len,
                   output_dim=embedding_size))
model.add(LSTM(50))
model.add(Dense(2,activation='softmax'))
from tensorflow.keras.optimizers import Adam
adam =Adam(learning_rate=0.01)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=8)
```

```python
#getting model performance metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, roc_auc_score
```

```python
y_pred = np.argmax(model.predict(x_test), axis=-1)
y_pred
```

```python
y_test[:,1]
```

```python
tab=confusion_matrix(y_pred,y_test[:,1])
tab
```

```python
tab.diagonal().sum()/tab.sum()
```
